```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = pd.DataFrame({
    'Age': [22, 25, 30, 35, 40, 45, 50, 55, 60],
    'Income': [2500, 3000, 3200, 4000, 5000, 5500, 6500, 7000, 7500]
})
```

```python
# Create a scatter plot
sns.scatterplot(data=data, x='Age', y='Income', color='blue', marker='o')
plt.title("Age vs. Income")
plt.show()
```

```python
# Create a histogram
sns.histplot(data['Income'], bins=10, color='green', kde=True)
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```

```python
# Create a box plot
sns.boxplot(data=data, y='Income', color='orange')
plt.title("Income Distribution - Box Plot")
plt.ylabel("Income")
plt.show()
```

```python
# Sample data
data = sns.load_dataset('iris')

# Create a pair plot
sns.pairplot(data, hue='species')
plt.suptitle("Pair Plot for Iris Dataset", y=1.02)
plt.show()
```

```python
# Correlation matrix
correlation_matrix = data.corr()

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(tips['total_bill'], kde=True, color='blue')
plt.title("Total Bill Distribution")

plt.subplot(1, 2, 2)
sns.histplot(tips['tip'], kde=True, color='green')
plt.title("Tip Distribution")

plt.tight_layout()
plt.show()
```

```python
# Scatter plot of Total Bill vs Tip
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', 
                style='smoker', palette='coolwarm')
plt.title("Total Bill vs Tip")
plt.show()
```

```python
# Box plot of Tip by Day
sns.boxplot(x='day', y='tip', data=tips, palette='muted')
plt.title("Tip Distribution by Day")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("hr_attrition_employee")

plt.figure(figsize=(10, 6))
sns.violinplot(x="Attrition", y="MonthlyIncome", data=data, palette="muted")

plt.title("Distribution of Monthly Income by Attrition", fontsize=16)
plt.xlabel("Attrition", fontsize=12)
plt.ylabel("Monthly Income", fontsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data for demonstration
data = pd.DataFrame({
    'Attrition': ['Yes'] * 100 + ['No'] * 100,
    'MonthlyIncome': sns.load_dataset('penguins').dropna()['bill_length_mm'].sample(200, replace=True) * 100
})

# Create the violin plot
plt.figure(figsize=(5, 5))
sns.violinplot(x='Attrition', y='MonthlyIncome', data=data, palette="muted")

# Add labels for annotation
plt.title("Labeled Violin Plot: Monthly Income by Attrition", fontsize=16, weight='bold')
plt.xlabel("Attrition", fontsize=12)
plt.ylabel("Monthly Income", fontsize=12)

# Annotating key parts of the plot
plt.text(-0.15, 7500, "Wider Section = High Density\n(More Employees)", fontsize=10, color="blue")
plt.text(0.85, 11500, "Narrow Section = Low Density\n(Fewer Employees)", fontsize=10, color="orange")
plt.text(0, 18000, "White Dot = Median", fontsize=10, color="black")
plt.text(0, 16500, "Black Bar = Interquartile Range (IQR)", fontsize=10, color="black")
plt.text(-0.45, 4000, "Low Income, High Attrition", fontsize=10, color="blue", weight='bold')
plt.text(0.6, 20000, "Higher Income, Low Attrition", fontsize=10, color="orange", weight='bold')

# Display the labeled violin plot
plt.tight_layout()
plt.show()
```
