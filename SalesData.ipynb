```python
import pandas as pd
```

```python
OrderbreakD = pd.read_excel('orderBreakdown.xlsx')
SalesOrder = pd.read_excel('salesOrder.xlsx')
```

```python
print('OrderBreakDown', OrderbreakD.shape)
OrderbreakD.head(3)
```

```python
print('ListofOrders', SalesOrder.shape)
SalesOrder.head(3)
```

```python
salesData = pd.merge(SalesOrder,OrderbreakD, on='Order ID', how='inner')
salesData.head(2)
```

```python
total_sales = salesData['Sales'].sum()
total_sales
```

```python
top_category = salesData.groupby('Category')['Sales'].sum().idxmax()
print(f"Top Product Category: {top_category}")
```

```python
region_profit_margin = salesData.groupby('Region').agg({'Profit': 'sum', 
                                                        'Sales': 'sum'})
region_profit_margin['Profit Margin'] = (region_profit_margin['Profit'] / 
                                         region_profit_margin['Sales']) * 100
print(region_profit_margin)
```

```python
discount_impact = salesData.groupby(['Category', 'Discount']).agg({'Profit': 'mean'})
print(discount_impact)
```

```python
top_city = salesData.groupby('City')['Sales'].sum().idxmax()
print(f"City with Highest Revenue: {top_city}")
```
