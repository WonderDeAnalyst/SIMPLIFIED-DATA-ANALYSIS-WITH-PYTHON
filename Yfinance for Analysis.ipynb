```python
# Fetch data for Bitcoin (BTC-USD)
bitcoin = yf.Ticker("BTC-USD")

# Get the last 6 months of historical data
bitcoin_data = bitcoin.history(period="6mo")

# Display the first few rows of the data
print(bitcoin_data.head())

# Plot Bitcoin's closing prices
bitcoin_data['Close'].plot(figsize=(10, 6), title="Bitcoin Closing Prices")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.grid()
plt.show()
```

```python
# Fetch data for Tesla (TSLA), Google (GOOGL), and Bitcoin (BTC-USD)
assets = ['TSLA', 'GOOGL', 'BTC-USD']
data = yf.download(assets, start='2020-01-01', end='2023-12-31')

# Display the first few rows
print(data.head())
```

```python
# Fetch Bitcoin data from 2020-01-01 to 2024-12-30
btc_data = yf.download('BTC-USD', start='2020-01-01', end='2024-12-31')
btc_data.tail()
```

```python
btc_data.isnull().sum()
```

```python
# Extract Adjusted each prices and handle missing values
btc_close = btc_data['Close'].fillna(method='ffill')
btc_high = btc_data['High'].fillna(method='ffill')
btc_low = btc_data['Low'].fillna(method='ffill')
btc_open = btc_data['Open'].fillna(method='ffill')
btc_volume = btc_data['Volume'].fillna(method='ffill')
# Display the cleaned data
btc_data.tail()
```

```python
plt.figure(figsize=(6, 6))
btc_close.plot(title='Bitcoin Adjusted Closing Prices', color='blue')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.show()
```

```python
# Calculate daily returns
btc_daily_returns = btc_close.pct_change()

# Plot daily returns
plt.figure(figsize=(10, 6))
btc_daily_returns.plot(title='Bitcoin Daily Returns', color='blue')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.show()

# Display basic statistics
print(btc_daily_returns.describe())
```

```python
# Calculate 50-day and 200-day moving averages
btc_ma50 = btc_close.rolling(window=50).mean()
btc_ma200 = btc_close.rolling(window=200).mean()

# Plot moving averages with price
plt.figure(figsize=(10, 6))
plt.plot(btc_close, label='Bitcoin Price', color='orange')
plt.plot(btc_ma50, label='50-Day MA', linestyle='--', color='blue')
plt.plot(btc_ma200, label='200-Day MA', linestyle='--', color='green')
plt.title('Bitcoin Price with Moving Averages')
plt.legend()
plt.show()
```

```python
# Calculate 30-day rolling standard deviation
btc_volatility = btc_daily_returns.rolling(window=30).std()

# Plot volatility
plt.figure(figsize=(10, 6))
plt.plot(btc_volatility, label='30-Day Rolling Volatility', color='red')
plt.title('Bitcoin Rolling Volatility')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.show()
```

```python
# Calculate daily returns for all assets
all_daily_returns = data.pct_change()

# Correlation matrix
correlation = all_daily_returns.corr()
print(correlation)

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Download data for EUR/USD and GBP/USD
forex_pairs = ['EURUSD=X', 'GBPUSD=X']
forex_data = yf.download(forex_pairs, start='2020-01-01', end='2024-12-31')

# Extract Adjusted Close prices
forex_close = forex_data['Close']

# Display the first few rows
print(forex_close.head())
```

```python
plt.figure(figsize=(10, 6))
forex_close.plot(title='Forex Exchange Rates: EUR/USD and GBP/USD')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.legend(forex_close.columns)
plt.show()
```

```python
# Calculate daily returns
forex_returns = forex_close.pct_change()

# Plot daily returns
plt.figure(figsize=(10, 6))
forex_returns.plot(title='Daily Returns: EUR/USD and GBP/USD')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.legend(forex_returns.columns)
plt.show()

# Display basic statistics of daily returns
print(forex_returns.describe())
```

```python
# Calculate the correlation matrix
correlation = forex_returns.corr()
print("Correlation Matrix:\n", correlation)

# Visualize correlation with a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Between EUR/USD and GBP/USD')
plt.show()
```
