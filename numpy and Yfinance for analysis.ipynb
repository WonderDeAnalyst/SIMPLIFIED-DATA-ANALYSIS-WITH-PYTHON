```python
from numpy.fft import fft

# Generate a signal
t = np.linspace(0, 1, 500)
signal = np.sin(2 * np.pi * 50 * t) + np.sin(2 * np.pi * 120 * t)

# Compute the Fourier Transform
freq_domain = fft(signal)
print("Frequency domain signal:", freq_domain[:10])
```

```python
# Monte Carlo simulation for estimating pi
num_samples = 1000000
x = np.random.rand(num_samples)
y = np.random.rand(num_samples)

# Count points inside the unit circle
inside_circle = (x**2 + y**2) <= 1
pi_estimate = (inside_circle.sum() / num_samples) * 4
print("Estimated value of pi:", pi_estimate)
```

```python
# Calculate daily returns
prices = np.array([100, 101, 103, 102, 104])
returns = (prices[1:] - prices[:-1]) / prices[:-1]
print("Daily returns:", returns)
```

```python
# Implement gradient descent
def gradient_descent(x, grad, lr=0.1, iterations=100):
    for _ in range(iterations):
        x -= lr * grad(x)
    return x

# Gradient of x^2
grad = lambda x: 2 * x
result = gradient_descent(np.array([10.0]), grad)
print("Minimized value:", result)
```

```python
import numpy as np
import pandas as pd

# Create a DataFrame from a NumPy array
data = np.random.rand(100, 3)  # 100 rows, 3 columns
columns = ['Feature1', 'Feature2', 'Feature3']
df = pd.DataFrame(data, columns=columns)

# Add a new column using NumPy operations
df['Sum'] = np.sum(df, axis=1)
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot
plt.plot(x, y, label='Sine Wave')
plt.title('Sine Wave')
plt.xlabel('X')
plt.ylabel('sin(X)')
plt.legend()
plt.show()
```

```python
# Aggregate data for visualization
mean_values = df.mean()

# Plot
mean_values.plot(kind='bar', color=['red', 'green', 'blue', 'orange'])
plt.title('Mean of Each Column')
plt.ylabel('Mean Value')
plt.show()
```

```python
# Generate synthetic time series data
dates = pd.date_range(start='2023-01-01', periods=100)
prices = np.cumsum(np.random.randn(100)) 
df = pd.DataFrame({'Date': dates, 'Price': prices})

# Plot the time series
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Price'], label='Stock Price', color='blue')
plt.title('Stock Price Trend')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Correlation matrix
corr_matrix = df.corr()

# Heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
